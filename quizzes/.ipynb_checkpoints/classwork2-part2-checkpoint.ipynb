### 6. (10 points) 2D Laplace equation: Electrostatic potential

We seek to find the electric potential for all the points inside a charge-free square with dimensions ($0.4\times 0.4 \,\rm m^2$). The top and sides of the region are made up of wires that are grounded (i.e. they are kept at $0\,\rm V$). The bottom wire is connected to a battery that keeps it at a constant $240\,\rm V$ (see figure below).

To find the potential, $V(x,y)$, we need to solve the 2D Laplace equation:

$$\frac{\partial^2 V}{\partial x^2} + \frac{\partial^2 V}{\partial y^2} = 0$$


(a) Write down the central-difference approximation of the Laplace equation, and indicate the expected errors. Hint: Express the answer with $V_{i,j}$ on the left-hand side, and justify all steps to get full marks.


(b) Create an empty matrix for $V(x,y)$ and add the boundary conditions of the problem to it. Hint: A matrix with $100\times 100$ cells would be sufficient.


(c) Create a method that iteratively fills up the solution using the finite-difference formula that you found in item (a). Hint: 1000 iterations should be sufficient to capture the surface adequately.

(d) Make a two-panel plot showing the solution for $V(x,y)$, as a surface (on left-hand side plot) and with contours projected onto the XY plane (on right-hand side plot).


(e) Compute the force field associated with the resulting potential and make a quiver map showing them on the XY plane. In which direction would a charged particle move?
